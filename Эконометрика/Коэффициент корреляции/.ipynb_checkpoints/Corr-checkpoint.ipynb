```python
import numpy as np
import pandas as pd
df = pd.read_excel('Dannye.xlsx', sheet_name='Данные для python')
```

```python
n = df.shape[0]
xmean=df['X'].mean()
ymean=df['Y'].mean()
#сумма x
x = df['X'].sum()
#сумма y
y = df['Y'].sum()
#сумма x^2
x2 = (df['X']*df['X']).sum()
#сумма y^2
y2 = (df['Y']*df['Y']).sum()
#сумма x*y
xy = (df['X']*df['Y']).sum()
#сумма x-xmean
x_xmean = (df['X'] - xmean).sum()
#сумма y-ymean
y_ymean = (df['Y'] - ymean).sum()
#сумма (x-xmean)^2
x_xmean2 = ((df['X'] - xmean)*(df['X'] - xmean)).sum()
#сумма (y-ymean)^2
y_ymean2 = ((df['Y'] - ymean)*(df['Y'] - ymean)).sum()
#сумма (x-xmean)*(y-ymean)
sumxymean = ((df['X'] - xmean)*(df['Y'] - ymean)).sum()
```

```python
#1
k1 = sumxymean / (x_xmean2*y_ymean2)**(1/2)
k1
```

```python
#2
k2 = (n*xy-x*y)/((n*x2-x*x)*(n*y2-y*y))**(1/2)
k2
```
