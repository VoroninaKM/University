```python
import numpy as np
import pandas as pd
y_t = pd.read_excel('Dannye.xlsx', sheet_name='Данные для python', index_col=0)
lag = int(input('Укажите номер лага: '))
y_tn = pd.DataFrame(y_t.iloc[:y_t.shape[0]-lag, 0], index=y_t.index).dropna()
y_t, y_tn
```

```python
n = y_tn.shape[0]

xmean=y_t.iloc[lag:,0].mean() #x - yt
ymean=y_tn.iloc[:,0].mean() #y - ytn

#сумма x-xmean
x_xmean = (y_t.iloc[lag:,0] - xmean).sum()
#сумма y-ymean
y_ymean = (y_tn.iloc[:,0] - ymean).sum()
#сумма (x-xmean)^2
x_xmean2 = ((y_t.iloc[lag:,0] - xmean)*(y_t.iloc[lag:,0] - xmean)).sum()
#сумма (y-ymean)^2
y_ymean2 = ((y_tn.iloc[:,0] - ymean)*(y_tn.iloc[:,0] - ymean)).sum()
#сумма (x-xmean)*(y-ymean)
sumxymean = ((y_t.iloc[lag:,0] - xmean).to_numpy()*(y_tn.iloc[:,0] - ymean).to_numpy()).sum()

print(sumxymean)
```
